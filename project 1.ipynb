## classify movie based on Ratings [Excellent , good or bad]

```python
def rating(rating):
    if rating>=7.0:
        return "Excellent"
    elif rating >=6.0:
        return "Good"
    else:
        return "Average"
```

```python
df["Rating_rates"] = df["IMDB_Rating"].apply(rating)
```

## Find unique values from Genre

```python
df["Genre"]
```

```python
list1 = []
for i in df["Genre"]:
    list1.append(i.split(","))
```

```python
list1
```

```python
one_d = []
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
uni_list=[]
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
```
